```python
num1 = int(input('digite um valor:'))
num2 = float(input('digite um valor:'))

operacao = input('digite a operação desejada: (+ , - , * , /)')

if operacao == '+':
  total = num1 + num2
  print(total)
elif operacao == '-':
  total = num1 - num2
  print(total)
elif operacao == '*':
  total = num1 * num2
  print(total)
elif operacao == '/':
  total = num1 / num2
  print(total)
```
